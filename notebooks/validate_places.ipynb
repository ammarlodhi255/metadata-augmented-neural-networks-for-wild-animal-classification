```python
import json
from PIL import Image
import numpy as np
import time
import matplotlib.pyplot as plt
import random
```

```python
def get_samples(bp):
    with open(bp + "metadata.json") as f:
        metadata = json.load(f)

    categories = metadata['categories']
    annotations = metadata['annotations']


    return categories, annotations
bp = "/media/user-1/CameraTraps/NINA/Images/"
categories, annotations = get_samples(bp)
```

```python
r1 = random.randint(0, len(annotations)-1)
r2 = random.randint(0, len(annotations)-1)
anot1 = annotations[r1]
anot2 = annotations[r2]
p1 = anot1['env_vector'][1:366]
p2 = anot2['env_vector'][1:366]
a1 = anot1['env_vector'][366:]
a2 = anot2['env_vector'][366:]

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.title("Scene recognition")
plt.plot(p1)
plt.plot(p2)
plt.subplot(122)
plt.title("Attributes")
plt.plot(a1)
plt.plot(a2)
```
